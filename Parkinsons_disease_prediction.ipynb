```python
x = data.drop(['name','status'] , axis=1)
y = data['status']
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.8, random_state=10)
```

```python
print(x.shape , x_train.shape , x_test.shape)  
```

```python
print(y.shape , y_train.shape , y_test.shape) 
```

```python
scaler = StandardScaler()
print(scaler.fit(x_train))
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

print(x_train)
print(x_test)
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(x_train , y_train)
```

```python
x_train_prediction = model.predict(x_train)
print(confusion_matrix(x_train_prediction,y_train))

# Accuracy Score On Training Data

accuracy = accuracy_score(y_train, x_train_prediction)
print('\n Accuracy on train data :' , accuracy)

# Errors On Training Data

MAE = metrics.mean_absolute_error(x_train_prediction, y_train)
MSE = metrics.mean_squared_error(x_train_prediction, y_train)                       
print('\n Mean Absolute Error on train data :', MAE)
print('Mean Squared Error on train data :', MSE)
```

```python
x_test_prediction = model.predict(x_test)
print(confusion_matrix(x_test_prediction,y_test))

# Accuracy Score On Test Data

accuracy = accuracy_score(y_test, x_test_prediction)
print('\n Accuracy on test data :' , accuracy)

# Errors on Test Data

MAE = metrics.mean_absolute_error(x_test_prediction, y_test)
MSE = metrics.mean_squared_error(x_test_prediction, y_test)
                                
print('\n Mean Absolute Error on test data:', MAE)
print('Mean Squared Error on test data :', MSE)
```
